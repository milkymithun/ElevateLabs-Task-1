### Data Cleaning

```python
print("No null values, no need of handling them")
```

```python
# Converting the mismatched data types into right data types

print(medical_df['PatientId'].dtype)

print("\n")

medical_df['PatientId'] = medical_df['PatientId'].astype(int)


print(medical_df["PatientId"].dtype)
```

```python
# As PatientId is in scientific representation, though to represent in exact value

# no use about it, just for clarification

pd.set_option('display.float_format', '{:.0f}'.format)
```

```python
medical_df['ScheduledDay']


print("""

This is the Common Timezone called UTC(Corordinated Universal Timezone),through which we can convert the time into India,Eatsern and Pacific

1. India(IST)= UTC + 5:30
2. Eastern(EDT) = UTC - 4:00
3. Pacific(PDT) = UTC - 7:00



""")

from datetime import datetime

medical_df['ScheduledDay'] = pd.to_datetime(medical_df['ScheduledDay'], utc = True)      # utc parameter ensures this is a UTC
medical_df['ScheduledDay'] = medical_df['ScheduledDay'].dt.tz_convert('Asia/Kolkata')      # Asia/Kolkata for IST

medical_df['ScheduledDay']
```

```python
# similarly convert for AppointmentDay attribute

medical_df['AppointmentDay'] = pd.to_datetime(medical_df['AppointmentDay'], utc = True)
medical_df['AppointmentDay'] = medical_df['AppointmentDay'].dt.tz_convert('Asia/Kolkata')

medical_df['AppointmentDay']
```

```python
# Extracting only dates from the ScheduledDay and AppointmentDay

medical_df['ScheduledDay'] = medical_df['ScheduledDay'].dt.date

medical_df['AppointmentDay'] = medical_df['AppointmentDay'].dt.date

print(medical_df['ScheduledDay'])

print(medical_df['AppointmentDay'])



```

```python
medical_df['AppointmentDay'] = pd.to_datetime(medical_df['AppointmentDay'])
medical_df['ScheduledDay'] = pd.to_datetime(medical_df['ScheduledDay'])

# converting from object datatype to datetime
```
